## **Feature Engineering**

```python
# Import RandomOverSampler from imblearn library
from imblearn.over_sampling import RandomOverSampler
ros = RandomOverSampler(random_state=100)
# Perform resampling on training data
X_resampled, y_resampled = ros.fit_resample(X_train, y_train)
# Print shapes
print(f'Original training data shape: {X_train.shape}, {y_train.shape}\n')
print(f'Resampled training data shape: {X_resampled.shape}, {y_resampled.shape}\n')

print(f'Class distribution in original training data: {y_train.value_counts(normalize=True)}\n')
print(f'Class distribution in resampled training data: {y_resampled.value_counts(normalize=True)}\n')
```

Using RandomOverSampler to address class imbalance in the training set. Fraudulent claims increased from 25% to 50% after resampling. That helped improve model performance on the validation set, particularly in terms of recall and F1-score. But it also led to some overfitting, as the model performed well on the training set but struggled with generalization to the validation set.

```python
# Create new features based on your understanding for both training and validation data
def create_new_features(df):
    df = df.copy()
    # Feature 1: gap between incident and policy bind dates
    df['incident_policy_gap'] = (df['incident_date'] - df['policy_bind_date']).dt.days
    # Feature 2: claim amount per vehicle
    df['claim_amount_per_vehicle'] = df['total_claim_amount']/df['number_of_vehicles_involved'].replace(0, 1)
    # Feature 3: annual premium to claim amount ratio
    df['premium_to_claim_ratio'] = df['policy_annual_premium']/df['total_claim_amount'].replace(0, 1)
    # Feature 4: injury to property claim ratio
    df['injury_to_property_claim_ratio'] = df['injury_claim']/df['property_claim'].replace(0, 1)
    # Feature 5: binary capital loss indicator
    df['is_capital_loss'] = (df['capital-loss'] < 0).astype(int)
    # Feature 6: binary capital gain indicator
    df['is_capital_gain'] = (df['capital-gains'] > 0).astype(int)
    # Feature 7: rrdinal indicator for incident_severity
    severity_mapping = {'Trivial Damage': 0, 'Minor Damage': 1, 'Major Damage': 2, 'Total Loss': 3}
    df['incident_severity_encoded'] = df['incident_severity'].map(severity_mapping)
    # Feature 8: interaction incident_severity_encoded with total_claim_amount
    df['severity_claim_interaction'] = df['incident_severity_encoded']*df['total_claim_amount'].replace(0, 1)
    return df
X_resampled = create_new_features(X_resampled)
X_val = create_new_features(X_val)
```

Created new features such as 'incident_policy_gap', 'claim_amount_per_vehicle', 'premium_to_claim_ratio', 'injury_to_property_claim_ratio', 'is_capital_loss', 'is_capital_gain', 'incident_severity_encoded',  'incident_severity_encoded', and 'severity_claim_interaction' to capture temporal aspects of the claims. 
- 'incident_policy_gap': The time difference between the policy start date and the incident date,
- 'claim_amount_per_vehicle': The total claim amount divided by the number of vehicles involved in the incident,
- 'premium_to_claim_ratio': The ratio of the annual premium to the total claim amount,
- 'injury_to_property_claim_ratio': The ratio of the total injury claim amount to the total property claim amount,
- 'is_capital_loss': A binary feature indicating whether the claim involves capital loss,
- 'is_capital_gain': A binary feature indicating whether the claim involves capital gain,
- 'incident_severity_encoded': The incident severity encoded as a numerical feature,
- 'severity_claim_interaction': An interaction feature between incident severity and claim amount.

```python
# Combine categories that have low frequency or provide limited predictive information

# Function to combine categories based on frequency and fraud likelihood
def combine_categories(df_train, df_val, y_train, col, freq_threshold=0.05, likelihood_threshold=0.247):
    df_train = df_train.copy()
    df_train['target'] = y_train
    df_val = df_val.copy()
    # Calculate frequency
    freq = df_train[col].value_counts(normalize=True)
    # Identify low frequency categories
    low_freq_categories = freq[freq < freq_threshold].index.tolist()
    # Calculate fraud likelihood
    fraud_likelihood = df_train.groupby(col)['target'].mean()
    # Identify low likelihood categories
    low_likelihood_categories = fraud_likelihood[fraud_likelihood < likelihood_threshold].index.tolist()
    # Combine low frequency or low likelihood categories
    combined_low_categories = set(low_freq_categories + low_likelihood_categories)
    df_train[col] = df_train[col].apply(lambda x: 'Other' if x in combined_low_categories else x)
    df_val[col] = df_val[col].apply(lambda x: 'Other' if x in combined_low_categories else x)
    # Drop the target column from training data
    df_train = df_train.drop(columns=['target'])
    return df_train, df_val
# Check the number of unique categories before combining
X_train_nunique_before = X_resampled_handled[categorical_cols].nunique()
X_val_nunique_before = X_val_handled[categorical_cols].nunique()
# Apply the function to categorical columns in training and validation datasets

for col in categorical_cols:
    X_resampled_handled, X_val_handled = combine_categories(X_resampled_handled, X_val_handled, y_resampled, col)

X_train_nunique_after = X_resampled_handled[categorical_cols].nunique()
X_val_nunique_after = X_val_handled[categorical_cols].nunique()

print('Resampled training set shape:', X_resampled_handled.shape)
print('Resampled validation set shape:', X_val_handled.shape)

print(f'\nNumber of unique categories in training before combining:\n {X_train_nunique_before}')
print(f'\nNumber of unique categories in training after combining:\n {X_train_nunique_after}')

print(f'\nNumber of unique categories in validation before combining:\n {X_val_nunique_before}')
print(f'\nNumber of unique categories in validation after combining:\n {X_val_nunique_after}')
```

```python
# Import libraries or function to plot the ROC curve
from sklearn.metrics import roc_curve, roc_auc_score

# Define ROC function
def plot_roc_curve(y_true, y_scores):
    fpr, tpr, thresholds = roc_curve(y_true, y_scores)
    auc_lr = roc_auc_score(y_true, y_scores)
    plt.figure(figsize=(12, 8))
    plt.plot(fpr, tpr, color='blue', label=f'ROC curve (area = {auc_lr:.4f})')
    plt.plot([0, 1], [0, 1], color='red', linestyle='--')
    plt.xlim([0.0, 1.0])
    plt.ylim([0.0, 1.05])
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title('Receiver Operating Characteristic (ROC) Curve')
    plt.legend(loc='lower right')
    plt.show()
    plt.close()
```

ROC curve helps to visualise the trade-off between sensitivity and specificity across different classification thresholds. ### **Hyperparameter Tuning**

```python
# Use grid search to find the best hyperparamter values
parameter_grid = {
    'n_estimators': [100, 200, 300],
    'max_depth': [5, 6, 8],
    'min_samples_split': [25, 30],
    'min_samples_leaf': [20, 25],
    'max_features': [0.2, 0.3, 0.4],
    'bootstrap': [True, False]
}
cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=100)
grid_search = GridSearchCV(estimator=RandomForestClassifier(random_state=100, class_weight='balanced_subsample'),
                           param_grid=parameter_grid,
                           cv=cv,
                           scoring='f1_weighted',
                           n_jobs=-1,
                           verbose=2)
# Best Hyperparameters
grid_search.fit(X_train_rf, y_resampled)
best_params = grid_search.best_params_
print('Best hyperparameters:', best_params)
```

The best hyperparameters found through grid search are as follows: 'bootstrap': False, 'max_depth': 8, 'max_features': 0.4, 'min_samples_leaf': 20, 'min_samples_split': 25, 'n_estimators': 100, 'class_weight': 'balanced_subsample', and cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=100). These hyperparameters aim to improve the model's performance on the validation set and address the class imbalance issue.

```python
# Building random forest model based on results of hyperparameter tuning
rf_model_tuned = RandomForestClassifier(**grid_search.best_params_, random_state=100, class_weight='balanced_subsample', n_jobs=-1)
rf_model_tuned.fit(X_train_rf, y_resampled)
```
